## Data Cleaning

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()

df = df.drop_duplicates()
```

```python
df['calculated_revenue'] = df['discounted_price'] * df['quantity_sold']
```
