## Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/cleaned_sales_data.csv")
```

### Revenue Overview

```python
df['total_revenue'].sum()
```

```python
df['order_id'].nunique()
```

```python
df['quantity_sold'].sum()
```

```python
category_revenue = df.groupby('product_category')['total_revenue'].sum().sort_values(ascending=False)

category_revenue.plot(kind='bar', figsize=(10,5))

plt.title("Revenue by Product Category")
plt.show()
```

```python
region_sales = df.groupby('customer_region')['total_revenue'].sum()

region_sales.plot(kind='bar', figsize=(10,5))

plt.title("Revenue by Region")
plt.show()
```

```python
sns.scatterplot(
    x=df['discount_percent'],
    y=df['quantity_sold']
)

plt.title("Discount Impact on Sales")
plt.show()
```

```python
sns.scatterplot(
    x=df['rating'],
    y=df['quantity_sold']
)

plt.title("Ratings vs Sales")
plt.show()
```
